```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from numpy import random
import pandas as pd
```

```python
class Data:
    def __init__(self, size, noise_var, noise_mean, a, b, outliers_num = 0, outliers_mag = 0):
        self.size = size
        self.x    = np.random.randint(low = 0, high = 10, size =self.size).reshape(-1,1)
        noise   = random.normal(loc = 0,scale = noise_var, size = self.size).reshape(-1,1)
        
        x1     = (self.x * a[0]) + b[0]
        x2     = (self.x * a[1]) + b[1]
        x3     = (self.x * a[2]) + b[2]
        x4     = (self.x * a[3]) + b[3]
        x5     = (self.x * a[4]) + b[4]
        x6     = (self.x * a[5]) + b[5]
        x7     = (self.x * a[6]) + b[6]
        x8     = (self.x * a[7]) + b[7]
        x9     = (self.x * a[8]) + b[8]
        x10    = (self.x * a[9]) + b[9]
        self.x = np.hstack((x1,x2,x3,x4,x5,x6,x7,x8,x9,x10))
        self.y = (a[0]*x1 + a[1]*x1 + a[2]*x1 + a[3]*x1 + a[4]*x1 +
                  a[5]*x1 + a[6]*x1 + a[7]*x1 + a[8]*x1 + a[9]*x1) + noise
        
        if outliers_num == 0:
            pass
        else:
            out = np.random.randint()
            for i in range(outliers_num):
                pass
                
```
