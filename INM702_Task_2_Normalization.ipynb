```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from numpy import random
import pandas as pd
import sklearn.preprocessing as pre
```

```python
# Creating covariates with different probability distribution

np.random.seed(42)

noise = random.normal(loc = 0, scale = 10, size = 100).reshape(-1,1)
approx_coefs = np.random.rand(3).reshape(3,1)
inter = np.random.rand(1)

c1 = random.normal(loc = 0, scale = 5, size = 100).reshape(-1,1)
c2 = 100 * random.gamma(shape = 10, scale = 10, size = 100).reshape(-1,1)
c3 = 20 * random.uniform(low = 0,high = 100, size = 100).reshape(-1,1)

print('approx coefs:\n', approx_coefs)

X = np.hstack((c1,c2,c3))
y = np.dot(X , approx_coefs) + inter + noise * 10


data = pd.DataFrame(data = np.hstack((X,y)))
data.columns = ['c1','c2','c3','y']
data
```

```python
# Using regression model without normalization

X = np.hstack((c1,c2,c3))
y = np.dot(X , approx_coefs) + noise * 20


Model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
Model.fit(X_train,y_train)
print(Model.coef_)
print(Model.score(X_train, y_train))
print(Model.rank_)

data = pd.DataFrame(data = np.hstack((X,y)))
data.columns = ['c1','c2','c3','y']
data
```

```python
X = np.hstack((c1,c2,c3))
y = np.dot(X , approx_coefs) + inter + noise * 20


# min max normalization
S = pre.MinMaxScaler()
X = S.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

Model = LinearRegression()
Model.fit(X_train,y_train)
print(Model.coef_)
print(Model.score(X_train, y_train))
print(Model.rank_)

data = pd.DataFrame(data = np.hstack((X,y)))
data.columns = ['c1','c2','c3','y']
data
```
